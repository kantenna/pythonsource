```python
# 클래스 메소드

class SelfTest:
    @classmethod
    def func1(cls): # 클래스 메소드
        print("function1 called")

    def func2(self): # 인스턴스 메소드
        print("function2 called")
```

```python
class SelfTest:

    num = 15

    @staticmethod
    def func1(a, b): # static 메소드 : 변수(클래스 변수, self) 사용 안함 / 외부 변수 사용
        return a + b

    def func2(self): # 인스턴스 메소드
        print("function2 called")

    @classmethod
    def func3(cls): # 클래스 메소드
        print(f"class method {cls.num}")
```

```python
test = SelfTest()
test.func2()

SelfTest.func3()
print(SelfTest.func1(3,5))
```

```python
class Parent():
    def __init__(self):
        self.value = "테스트"

    def test(self):
        print("Parent test() 호출")

class Child(Parent): # 상속
    def __init__(self):
        Parent.__init__(self) # == super() 부모의 인스턴스 변수 호출 여부
        print("Child 클래스의 __init__ 호출")
```

```python
parent = Parent()
parent.test()

child = Child()
child.test()
print(child.value) # 부모의 변수 호출
```

```python
class AudioVisual:
    def __init__(self, power, volume):
        self.power = power
        self.volume = volume

    def switch(self,onOff):
        self.power = onOff

    def setVolume(self, volume):
        self.volume = volume


class Audio(AudioVisual):
    def __init__(self, power, volume):
        super().__init__(power, volume)

    def tune(self):
        str = "La la la ...." if self.power else  "turn it on"
        print(str)

class Tv(AudioVisual):
    def __init__(self, power, volume, size):
        super().__init__(power, volume)
        self.size = size

    def watch(self):
        str = "have fun !! " if self.power else "switch on"
        print(str)
```

```python
obj1 = Tv(False, 12, 40)
obj1.switch(True)
obj1.watch()

obj2 = Audio(False, 15)
obj2.setVolume(6)
obj2.tune()
```

```python
# 오버라이딩

class Parent():
    def __init__(self):
        self.value = "테스트"

    def func1(self):
        print("Parent func1() 호출")

    def func2(self):
        print("Parent func2() 호출")

class Child(Parent): # 상속
    def __init__(self):
        Parent.__init__(self) # == super() 부모의 인스턴스 변수 호출 여부
        print("Child 클래스의 __init__ 호출")

    def func1(self):
        print("Child func1() 재정의")
    
    def func3(self):
        print("Child func3() 호출")
```

```python
parent, child = Parent(), Child()
parent.func1()
parent.func2()

child.func1()
child.func2()
child.func3()
```

```python
# private 키워드는 없음
# __변수 : private으로 취급됨
import math
class Circle():
    def __init__(self, radius):
        self.__radius = radius

    def getcirc(self):
        return 2 * math.pi * self.__radius
    
    def getarea(self):
        return math.pi * (self.__radius ** 2)
    
    def getradius(self):
        return self.__radius
    
    def setradius(self, radius):
        self.__radius = radius
```

```python
circle = Circle(10)
print(f"둘레 {circle.getcirc()}")
print(f"면적 {circle.getarea()}")
print(circle.getradius)
```
