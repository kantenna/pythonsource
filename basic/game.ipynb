- word.txt 읽은 후 리스트 생성
- 리스트 섞은 후 단어를 하나 추출 => 추출된 단어 보여주기
- 사용자는 보여진 단어를 똑같이 타자치기
- 5개의 문제 제출, 단어 입력

단어를 4개 이상 정확히 입력하면 합격

게임 시간 재기

출력문
게임시간 : 3.05초, 정답개수 : 4개
합격

```python
import time
import random
```

```python
# 1. word.txt 읽어서 리스트 생성
words = []
with open("./data/word.txt", "r", encoding="utf-8") as fr:
    for c in fr:
        words.append(c.strip())
```

```python
input("게임 시작 엔터키 입력")
start = time.time()
n = 1
corr_cnt = 0 # 정답 갯수

while n <= 5:
    random.shuffle(words)
    q = random.choice(words)

    print()
    print("** Question #{n}")
    print(q) # 문제 보여주기
    x = input()
    n += 1

    # 문제와 사용자의 입력값 일치 여부 확인
    if x == q:
        corr_cnt += 1
        print("Pass!!")
    else:
        print("Wrong!!")

end = time.time()
# 걸린 시간
et = end - start

print("*** 게임 결과 ***")
print(f"게임시간:  {et:.3f}초, 정담 갯수 : {corr_cnt}")
if corr_cnt > 3:
    print("합격")
else:
    print("불합격")
```
