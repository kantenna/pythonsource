```python
# 동전 교환하기
# 7777 => 동전 500원, 100원, 50원, 10원
def coin_change(money, coins):
    count = 0
    result = {}

    for coin in coins:
        cnt = money // coin
        money %= coin
        count += cnt
        result[coin] = cnt

    return count, result


coins = [500, 100, 50, 10]
money = 7777

total, detail = coin_change(money, coins)

print("총 동전 개수:", total)
print("동전별 개수:", detail)
```

```python
# 사칙연산기 (숫자 2개 입력, 연산자 => +,-,*,/,//,%)

num1 = int(input('input num1 '))
op = input('input op ')
num2 = int(input('input num2 '))

if op == "+":
    result = num1 + num2
elif op == "-":
    result = num1 - num2
elif op == "*":
    result = num1 * num2
elif op == "/":
    result = num1 / num2
elif op == "//":
    result = num1 // num2
elif op == "%":
    result = num1 % num2

print(f"{num1} {op} {num2} = {result}")
```
