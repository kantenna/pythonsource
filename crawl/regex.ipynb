#### 정규표현식
- 정규식 패턴은 동일

```python
data = """\
park 800905-1049118
kim 700905-1059118\
"""
result = []

# 뒷자리를 ******* 출력

# 엔터를 기준으로 문자열 분리
words = data.split("\n")
print(words)

# 공백을 기준으로 문자열 분리
for line in words:
    words_result = []
    # person = line.strip().split(" ")
    for word in line.strip().split():
        if len(word) == 14 and word[:6].isdigit() and word[7:].isdigit():
            word = word[:6] + "-" + "*******"
        words_result.append(word)

    result.append(" ".join(words_result))

# 리스트를 문자열 형태로 출력하는 방법
print("\n".join(result))
```
