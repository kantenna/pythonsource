```python
url = "https://jsonplaceholder.typicode.com/users"

with requests.Session() as s:

    res = s.get(url)
    contents = res.json() # python json == > dict

    for row in contents:
        for k,v in row.items():
            print(f"{k} : {v}", end=" ")
        print()
```
